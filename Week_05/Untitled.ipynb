```python
cover = Image.open('bookcover2.jpg')
cover_large =  ImageOps.scale(cover, 3)
cover_large
```

```python
grayscale = ImageOps.grayscale(cover_large)
grayscale
```

```python
enhancer = ImageEnhance.Contrast(grayscale)
enhanced = enhancer.enhance(1.2)
enhanced
```

```python
invert = ImageOps.invert(enhanced)
invert
```

```python
pytesseract.image_to_string(invert)
```
