### Convert dataframes to lonf format in preparation for merging

```python
sfert = fert.stack()
slife = life.stack()
spop = pop.stack()
```

### Create dataframe from all three datasets in long format

```python
d = {'fertility': sfert, 'lifeexp': slife, 'population': spop}
df2 = pd.DataFrame(d)
df2.head(12)
```

### Stack also the three columns to have only one column with values

```python
df3 = df2.stack()
df3
```

### Convert back to wide table with country and quantity as columns and year as indices

```python
df4 = df3.unstack((0,2))
df4
```

### Select three countries and plot them

```python
df4[['Germany', 'France', 'Sweden']].plot(figsize=(12,8))
```

### Convert table to have quantities as columns and country and year as indices and creat scatter plot if fertility vs life expectancy

```python
df5 = df3.unstack(2)
df5
```

```python
df5.plot.scatter('fertility', 'lifeexp', s=0.1, figsize=(12,8))
plt.title('With increasing life expectancy fertility decreases', fontsize=20)
plt.xlabel('Total fertility rate', fontsize=16)
plt.ylabel('Life expectancy / years', fontsize=16)
plt.tick_params(labelsize=14)
```

### Create scatter plot for specific year (1950)

```python
df6 = df3.unstack(1)
df6 = df6[1950]
df6 = df6.unstack(1)
df6
```

```python
df6.plot.scatter('fertility', 'lifeexp', s=10, figsize=(12,8))
```

Add different colours for each country

```python
cmap = plt.get_cmap('tab20', lut = len(df6)).colors
df6.plot.scatter('fertility', 'lifeexp', s=10, figsize=(12,8), c=cmap)
```

Make the circles represent the population size of each country

```python
c_size = df6['population']/100000
df6.plot.scatter('fertility', 'lifeexp', s=c_size, figsize=(12,8), c=cmap, alpha=0.7)
plt.title('1950', fontsize=20, y= 1.02)
plt.xlabel('Total fertility rate', fontsize=16)
plt.ylabel('Life expectancy [years]', fontsize=16)
plt.tick_params(labelsize=14)
```

### Create gif with plots for each year from 1960 to 2015

Prepare dataframe with year as coulmns

```python
df7 = df3.unstack((1,2))
df7
```

Loop through the years

```python
for i in range(1960,2016):
    df8 = df7[i]
    c_size = df8['population']/100_000
    fig = df8.plot.scatter('fertility', 'lifeexp', s=c_size, figsize=(12,8), c=cmap, alpha=0.7).get_figure()
    plt.axis((0, 10, 0, 100))
    plt.title(str(i))
    filename = 'figures/lifeexp_{}.png'.format(i)
    plt.savefig(filename)
    plt.close()
```

Create gif with imageio

```python
images = []

for i in range(1960,2016):
    filename = 'figures/lifeexp_{}.png'.format(i)
    images.append(imageio.imread(filename))
    
imageio.mimsave('gapminder.gif', images, fps=20)
```
