Outer join includes all countries weather they could be matched or not.

```python
merged_outer = pd.merge(life_exp, continents, left_on='Life expectancy', right_on='country', how='outer')
merged_outer
```

Let's stick with inner join for now. Otherwise, the data would have to be cleaned beforehand.

### So what's the mean of the life expectancy by continent in 2016?

```python
merged_inner.groupby('continent').mean()
```

### To remove rows or columns with missing values:
 - `df.dropna()`
     - `df.dropna(axis=0)` drop rows with missing data (default)
     - `df.dropna(axis=1)` drop columns with missing data
