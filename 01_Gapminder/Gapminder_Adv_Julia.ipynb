### Create gif with plots for each year from 1960 to 2015

Prepare dataframe with year as coulmns

```python

df9 = df2.reset_index()
df7 = pd.merge(df9, continents, left_on='level_0', right_on='country')
df7 = df7.set_index(['level_0', 'level_1'])
df7 = df7.stack()
df7 = df7.unstack((1,2))
df10 = df7[1960]
```

Loop through the years

```python
for i in range(1960,2016):
    df8 = df7[i]
    c_size = df8['population']/100_000
    sns.set(font_scale=1.6)
    fig = df8.plot.scatter('fertility', 'lifeexp', s=c_size, figsize=(12,8), c=cmap, alpha=0.6).get_figure()
    plt.axis((0, 10, 0, 100))
    plt.title(str(i), y= 1.02, fontsize = 30, weight='bold')
    plt.xlabel('Total fertility rate [children per woman]', labelpad = 10)
    plt.ylabel('Life expectancy [years]', labelpad = 10)
    filename = 'figures/lifeexp_{}.png'.format(i)
    plt.savefig(filename)
    plt.close()

from IPython.display import Image
Image(url='figures/lifeexp_2000.png')
```

## Some stuff on sizes of dots or circles

### Matplotlib is very smart. The area of the circles is directly related to the data you pass for the argument s. That makes the viewer percieve the size of the circles with the correct proportions. 
The circle at x=10 is about double the size of the circle at x=5.

```python
x = [5, 10, 15]
y = [1, 1, 1]
s = [500, 1000, 1500]

plt.figure(figsize=(12,4))
plt.scatter(x, y, s=s)
plt.axis((0, 20, 0, 2))
```

### In other plotting tools, the argument for a circle size is usually the radius. When the respective data is taken as the radius, the view will perceive the size of the circles incorrectly. Here is an example of the same diagram produced in R.
The circle at x=10 is about double the size of the circle at x=5.
In that case, the square root of the data should be passed as an argument.
